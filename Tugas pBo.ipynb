# Simple Hero Battle System for Beginners
This notebook demonstrates a simple hero battle system using basic Python OOP concepts.
We will create two heroes, Paquito and Tigreal, with basic attack and skill functionalities.

```python

class Hero:
    def __init__(self, name, hp, ad, skill):  # Corrected indentation
        # atribut hero
        self.name = name  # nama hero
        self.hp = hp      # (HP)
        self.ad = ad      # damage (AD)
        self.skill = skill  # Hero skill name

    def attack(self, target):

        print(f"{self.name} attacks {target.name}!")
        target.hp -= self.ad
        print(f"{target.name} loses {self.ad} HP.")

    def use_skill(self, target): #
        print(f"{self.name} uses skill {self.skill}!")
        skill_damage = self.ad * 2  # Skill damage is twice the regular attack
        target.hp -= skill_damage  # Target HP is reduced by skill damage
        print(f"{target.name} loses {skill_damage} HP.")

    def show_stats(self):

        print(f"{self.name} - HP: {self.hp}, AD: {self.ad}, Skill: {self.skill}")
```

```python
# buat Paquito
paquito = Hero("Paquito", 1200, 150, "Knockout Strike")

# statistik paquito
paquito.show_stats()
```

```python
# buat tigreal
tigreal = Hero("Tigreal", 1400, 100, "Implosion")

# statistik tigreal
tigreal.show_stats()
```

```python
paquito.attack(tigreal)
tigreal.show_stats()
paquito.use_skill(tigreal)
tigreal.show_stats()
```
